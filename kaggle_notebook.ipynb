```python
# Step 3 — Start Flask server in background
import subprocess, time, threading

def run_server():
    subprocess.run(['python', 'app.py'], check=True)

t = threading.Thread(target=run_server, daemon=True)
t.start()

print('Waiting for model to load (can take ~60 seconds)...')
time.sleep(60)
print('Server should be running!')
```

```python
# Step 4 — Expose via ngrok (free public URL)
# Get a free token from https://dashboard.ngrok.com/signup
from pyngrok import ngrok

NGROK_TOKEN = 'YOUR_NGROK_AUTH_TOKEN_HERE'  # paste your token here
ngrok.set_auth_token(NGROK_TOKEN)

tunnel = ngrok.connect(5000)
public_url = tunnel.public_url

print('='*60)
print(f'  Open NeuraTutor at: {public_url}')
print('='*60)
```
